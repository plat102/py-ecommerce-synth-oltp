```python
# test_cache.py
from mock_ecommerce.database.operations import get_existing_ids, execute_query
from mock_ecommerce.database.ddl import _CREATE_BRAND

# 1. Setup Data
print("Setup data...")
execute_query(_CREATE_BRAND)
execute_query("INSERT INTO brand (brand_name) VALUES ('Apple'), ('Samsung') ON CONFLICT DO NOTHING")

# 2. Test fetch db
print("\n--- Call 1 (DB Hit) ---")
ids_1 = get_existing_ids('brand', 'brand_id')
print(f"IDs: {ids_1}")
```
